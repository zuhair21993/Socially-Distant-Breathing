```python
NY_df = aqi_df.loc[aqi_df["State"] == "New York"]
```

```python
NY_df
```

```python
NY_2018_df = NY_df.loc[NY_df["Year"] == 2018]
```

```python
NY_2018_df
```

```python
NewYork_df = NY_2018_df.drop(['Days CO', 'Days NO2', 'Days Ozone', 'Days SO2', 'Days PM2.5', 'Days PM10'], axis=1)
```

```python
NewYork_df["Bad Days"] = NewYork_df["Unhealthy for Sensitive Groups Days"] + NewYork_df["Unhealthy Days"] + NewYork_df["Very Unhealthy Days"] + NewYork_df["Hazardous Days"]
```

```python
NewYork_df["Max_AQI"] = NewYork_df["Max AQI"]
```

```python
NewYork_df["90th_Percentile_AQI"] = NewYork_df["90th Percentile AQI"]
```

```python
NewYork_df["Median_AQI"] = NewYork_df["Median AQI"]
```

```python
NewYork_df
```

```python
NewYork_2018_df = NewYork_df.drop(['Unhealthy for Sensitive Groups Days', 'Unhealthy Days', 'Very Unhealthy Days', 'Hazardous Days', 'Max AQI', '90th Percentile AQI', 'Median AQI'], axis=1)
```

```python
NewYork_2018_df["% Good Days"]=NewYork_2018_df["Good Days"]/NewYork_2018_df["Days with AQI"]*100
```

```python
NewYork_2018_df["% Moderate Days"]=NewYork_2018_df["Moderate Days"]/NewYork_2018_df["Days with AQI"]*100
```

```python
NewYork_2018_df["% Bad Days"]=NewYork_2018_df["Bad Days"]/NewYork_2018_df["Days with AQI"]*100
```

```python
NewYork_2018_df
```
